```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
df = pd.read_csv("../Data/mimic/timings.txt")
```

```python
df.head()
```

```python
# change to minutes
df['querytime'] = df['querytime']/1000
df['querytime'] = df['querytime']/60
```

```python
df1 = df[df['queryid'] == 1]
df2 = df[df['queryid'] == 2]
df3 = df[df['queryid'] == 3]
df4 = df[df['queryid'] == 4]
df5 = df[df['queryid'] == 5]
df6 = df[df['queryid'] == 6]
df7 = df[df['queryid'] == 7]

# reset the indexes 
df1.index = pd.RangeIndex(len(df1.index))
df1.index = range(len(df1.index))
df2.index = pd.RangeIndex(len(df2.index))
df2.index = range(len(df2.index))
df3.index = pd.RangeIndex(len(df3.index))
df3.index = range(len(df3.index))
df4.index = pd.RangeIndex(len(df4.index))
df4.index = range(len(df4.index))
df5.index = pd.RangeIndex(len(df5.index))
df5.index = range(len(df5.index))
df6.index = pd.RangeIndex(len(df6.index))
df6.index = range(len(df6.index))
df7.index = pd.RangeIndex(len(df7.index))
df7.index = range(len(df7.index))
```

```python
# Dealing with font sizes
SMALL_SIZE = 15
MEDIUM_SIZE = 20
BIGGER_SIZE = 30
```

```python
figure(figsize=(15, 10), dpi=300)

plt.plot(np.arange(0,1100,100), df1["querytime"], label="Q1", marker="+", color="blue", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df2["querytime"], label="Q2", marker="x", color="darkorange", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df3["querytime"], label="Q3", marker=".", color="limegreen", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df4["querytime"], label="Q4", marker="^", color="crimson", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df5["querytime"], label="Q5", marker="v", color="mediumorchid", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df6["querytime"], label="Q6", marker="o", color="sienna", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df7["querytime"], label="Q7", marker="d", color="hotpink", ms=10, linewidth=2.5)

plt.xlabel('No. constraints')
plt.ylabel('Time (minutes)')

plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)   # fontsize of the figure title

plt.grid(color='grey', linestyle='-', linewidth=1)

plt.legend()

plt.savefig('Executiontime-vs-constraints.png')
```

```python
df1["slowdown"] = (df1.querytime)/df1.querytime[0]
df2["slowdown"] = (df2.querytime)/df2.querytime[0]
df3["slowdown"] = (df3.querytime)/df3.querytime[0]
df4["slowdown"] = (df4.querytime)/df4.querytime[0]
df5["slowdown"] = (df5.querytime)/df5.querytime[0]
df6["slowdown"] = (df6.querytime)/df6.querytime[0]
df7["slowdown"] = (df7.querytime)/df7.querytime[0]
```

```python
figure(figsize=(15, 10), dpi=300)

plt.plot(np.arange(0,1100,100), df1["slowdown"], label="Q1", marker="+", color="blue", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df2["slowdown"], label="Q2", marker="x", color="darkorange", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df3["slowdown"], label="Q3", marker=".", color="limegreen", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df4["slowdown"], label="Q4", marker="^", color="crimson", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df5["slowdown"], label="Q5", marker="v", color="mediumorchid", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df6["slowdown"], label="Q6", marker="o", color="sienna", ms=10, linewidth=2.5)
plt.plot(np.arange(0,1100,100), df7["slowdown"], label="Q7", marker="d", color="hotpink", ms=10, linewidth=2.5)

plt.xlabel('No. constraints')
plt.ylabel('Slowdown')

plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)   # fontsize of the figure title

plt.grid(color='grey', linestyle='-', linewidth=1)

plt.legend()

plt.savefig('Slowdown-vs-constraints-all.png')
```
